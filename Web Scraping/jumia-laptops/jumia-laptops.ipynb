```python
# import requests as req
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
```

```python
url = 'https://www.jumia.com.gh/laptops/'
# session = req.Session()
headers = {
#     "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
#     "Accept-Encoding": "gzip, deflate, br",
#     "Accept-Language": "en-US,en;q=0.9",
#     "Host": "httpbin.org",
    "User-Agent": "Mozilla/6.0"
}
```

```python
# response = session.get(url, headers=headers)
req = Request(url, headers=headers)
webpage = urlopen(req).read()
```

```python
page = soup(webpage, 'html.parser')
print(page.prettify())
```

```python
laptop_details = page.find('div', class_="-paxs row _no-g _4cl-3cm-shs") # find all laptop details on the page
print(laptop_details)
# laptop_names = [name.text.strip() for name in laptops_name_results ]
# print(len(laptop_details))

                   
```

```python
# laptop_cards = laptop_details.find_all('article', class_='prd _fb col c-prd') # This line is commented because you can't perform a find_all operation on it
laptop_names_and_details = laptop_details.find_all("h3", class_="name")
# print(laptop_names_and_details)
laptop_names = [names.text.strip() for names in laptop_names_and_details if names.text.strip() != '']
# print(laptop_names)
laptop_prices = [prices.text.strip() for prices in laptop_details.find_all("div", class_="prc") if prices.text.strip() != '']
# print(laptop_prices)

# laptop_prices = [price.text.strip() for price in laptop_price_results]
# print(len(laptop_prices))
# print(laptop_prices)
```

```python
new = []
names_and_prices = [(laptop_names[i],laptop_prices[i]) for i in range(len(laptop_names))]
print(names_and_prices)
```

```python
import pandas as pd
columns = ['Laptop', 'Price']
df = pd.DataFrame.from_records(data=names_and_prices, columns=columns)
print(df)
```

```python
df.to_csv("Jumia laptops.csv")
```
